```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
from geopy.geocoders import Nominatim
```

```python
# Initialize Nominatim API
geolocator = Nominatim(user_agent="MyApp")
```

```python
df_diffdisorders  = pd.read_csv("Resources/prevalence-by-mental-and-substance-use-disorder.csv")
```

```python
df_diffdisorders.head()
```

```python
def get_lat_lon(country):
    
    try:
        location = geolocator.geocode(country)
        return location.latitude, location.longitude
    except:
        return np.nan
```

```python
countries = df_diffdisorders ["Entity"].unique()
```

```python
countries_lat_lon = {}

for c in countries:
    print("country: ", c)
    countries_lat_lon[c] = get_lat_lon(c)
```

```python
countries_lat_lon
```

# 2nd CSV file

```python
df_diffdisorders["lat_lon"] = df_diffdisorders["Entity"].map(countries_lat_lon)
```

```python
df_diffdisorders.head(2)
```

```python
discard_indexes = df_diffdisorders[df_diffdisorders["lat_lon"].isnull()].index
df_diffdisorders = df_diffdisorders.drop(discard_indexes)
```

```python
df_diffdisorders.head(2)
```

```python
df_diffdisorders["latitude"] = df_diffdisorders["lat_lon"].map(lambda x: x[0])
df_diffdisorders["longitude"] = df_diffdisorders["lat_lon"].map(lambda x: x[1])
```

```python
df_diffdisorders
```

```python
df_diffdisorders = df_diffdisorders.drop('lat_lon', axis=1)
```

```python
df_diffdisorders.head(2)
```

```python
df_diffdisorders = df_diffdisorders.rename(columns = {
'Prevalence - Schizophrenia - Sex: Both - Age: Age-standardized (Percent)':'Schizophrenia',
'Prevalence - Bipolar disorder - Sex: Both - Age: Age-standardized (Percent)':'Bipolar_disorder',  
'Prevalence - Eating disorders - Sex: Both - Age: Age-standardized (Percent)':'Eating_disorder',                                                  
'Prevalence - Anxiety disorders - Sex: Both - Age: Age-standardized (Percent)':'Anxiety',
'Prevalence - Drug use disorders - Sex: Both - Age: Age-standardized (Percent)':'Drug_usage',
'Prevalence - Depressive disorders - Sex: Both - Age: Age-standardized (Percent)':'Depression',
'Prevalence - Alcohol use disorders - Sex: Both - Age: Age-standardized (Percent)':'Alcohol'})
                                                                                                                
```

```python
df_diffdisorders.drop('Code',axis=1,inplace=True)
```

```python
df_diffdisorders
```

```python
df_diffdisorders.to_csv('Clean_data/df_diffdisorders.csv', index=False)
```
